```python
import pyspark

from pyspark.sql import SQLContext
from pyspark.sql import SparkSession

import pandas as pd
```

```python
SOURCE_DATASET_FOLDER = "./results/CICIDS2017_fine/X_train.csv"
```

```python
def _validate_existed_datasets():
    pass


def read_dataset():
    """Read source dataset with sql interface"""
    spark = SparkSession \
    .builder \
    .appName("Spark SQL Reader") \
    .getOrCreate()
    
    spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(SOURCE_DATASET_FOLDER) 
    data = spark.read.option("header", "true").csv(SOURCE_DATASET_FOLDER)
    return data
```
